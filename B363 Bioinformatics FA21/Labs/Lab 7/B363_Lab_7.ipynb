```python
def find_breakpoints(arr):
  length = len(arr)

  arr.insert(0, 0)
  arr.append(length + 1)

  breaks = 0
  for i in range(length - 1):
    if arr[i] + 1 != arr[i+1]:
      breaks += 1
  return breaks

string = '+3 +4 +5 -12 -8 -7 -6 +1 +2 +10 +9 -11 +13 +14'
arr = [int(item) for item in string.split()]
bre = find_breakpoints(arr)
print(bre)
```

```python
#Construct the De Bruijn Graph of a String
f = open('/content/rosalind_ba3d.txt', 'r').readlines()
k = int(f[0])

dna = ''.join(f[1:])
print(dna)

kmers = []

for i in range(len(dna) - k):
  kmers.append(dna[i:i+k])

print(kmers)

graph = {}
for kmer in kmers:
  if kmer[:k-1] not in graph:
    graph[kmer[:k - 1]] = [kmer[1:]]
  else:
    graph[kmer[:k - 1]].append(kmer[1:])

lst = sorted(list(graph.items()), key=lambda x: x[0])
out = ''

for item in lst:
  key = item[0]
  val = sorted(item[1])
  out += f'{key} -> {val}\n'

out = out.replace(']', '')
out = out.replace('[', '')
out = out.replace('\'', '')
out = out.replace('n', '')
out = out.replace('\\', '')

print(out)
```

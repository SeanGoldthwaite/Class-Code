```python
import numpy as np

def find_lcs(string1, string2):
  m = len(string1)
  n = len(string2)
  arr = np.zeros((m + 1, n + 1))
  path = np.full((m + 1, n + 1), fill_value=None)
  for i in range(m + 1):
    for j in range(n + 1):
      if i == 0:
        path[i,j] = 'd'
      elif j == 0:
        path[i,j] = 'i'
      elif string1[i - 1] == string2[j - 1]:
        arr[i,j] = 1 + arr[i - 1, j - 1]
        path[i,j] = 'm'
      elif arr[i - 1, j] >= arr[i, j - 1]:
        arr[i, j] = arr[i - 1, j]
        path[i, j] = 'i'
      else:
        arr[i, j] = arr[i, j - 1]
        path[i, j] = 'd'

  print(arr, '\n', path)

  i = m
  j = n

  lcs = ''
  res1, res2 = '', ''

  while i > 0 and j > 0:
    #print(i, j, path[i,j], string1[i - 1], string2[j - 1])
    if path[i,j] == 'd':
      j -= 1
      res2 += '-'
    elif path[i,j] == 'i':
      i -= 1
      res1 += '-'
    elif path[i,j] == 'm':
      i -= 1
      j -= 1
      res1 += string1[i]
      res2 += string2[j]
      lcs += string1[i]

  # print(res1[::-1])
  # print(res2[::-1])
  # print(lcs[::-1])
  return lcs[::-1], arr[m,n]
```
